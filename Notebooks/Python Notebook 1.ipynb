**Data Tranformation**

```python
import pandas as pd
df = pd.read_csv('/5DATA002W.2 Coursework Dataset(25012025v6.0).csv')
```

**Handling Missing Values**

**Missing values are found under these features**
Regional_Node_Examined,
Tumor_Size,
Age,
Sex   

Handling missing values in SEX vaiable

```python
df['Sex'].unique()
```

```python
# Replace '1' and any representations of missing values (nan, NaN, None) with 'Female'
df['Sex'] = df['Sex'].replace({1: 'Female', '1': 'Female', float('nan'): 'Female'})

# Convert the column to a consistent string type
df['Sex'] = df['Sex'].astype(str)

# Ensure all 'female' entries are standardized to 'Female'
df['Sex'] = df['Sex'].str.strip().str.lower().replace({'female': 'Female'})
```

Handling missing values in Regional_Node_Examined

```python
df['Regional_Node_Examined'].unique()
```

Here using median value we replace the missing value in Regional_Node_Examined which is 41.

```python
df['Regional_Node_Examined'].fillna(df['Regional_Node_Examined'].median(), inplace=True)
df['Regional_Node_Examined'].unique()
```

Handling missing values in Tumor_Size

```python
df['Tumor_Size'].unique()
```

```python
import numpy as np
```

```python

# Calculate IQR
Q1 = df['Tumor_Size'].quantile(0.25)  # 25th percentile
Q3 = df['Tumor_Size'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mean_value = df['Tumor_Size'].mean()
df['Tumor_Size'] = np.where(df['Tumor_Size'] < lower_bound, mean_value, df['Tumor_Size'])

```

```python
print("Min:", df['Tumor_Size'].min())
print("Max:", df['Tumor_Size'].max())
```

Handling missing values in Age

```python
df['Age'].unique()
```

```python
#Calculate IQR
Q1 = df['Age'].quantile(0.25)  # 25th percentile
Q3 = df['Age'].quantile(0.75)  # 75th percentile
IQR = Q3 - Q1  # Interquartile range

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound", lower_bound)
print("Upper Bound", upper_bound)
print('Age above Upper Bound : ' , len(df[df['Age'] > upper_bound]))
print('Age below Lower Bound : ' , len(df[df['Age'] < lower_bound]))
```

```python
mean_value = df['Age'].mean()

#Replacing the outlier with mean value
df['Age'] = np.where(df['Age'] < lower_bound, mean_value, df['Age'])

#Replacing blank values with mean
df['Age'].fillna(df['Age'].median(), inplace=True)

print('Age below Lower Bound : ' , len(df[df['Age'] < lower_bound]))
```
